```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'E:\my project\datafile.csv')

```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
# Convert numerical columns to numeric data type, replacing non-numeric values with NaN
numeric_columns = df.columns[3:-1]  # Select columns with numeric data
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
# Group the data by sector and calculate the mean for each sector
grouped_data = df.groupby('Sector').mean()
```

```python
# Plot the average values for each sector
grouped_data.plot(kind='bar', figsize=(25,15))
plt.xlabel('Sector')
plt.ylabel('Average Value')
plt.title('Average Values by Sector')
plt.legend(loc='upper right')
plt.show()
```

```python
# Convert Year column to datetime type
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
```

```python
# Plot multiple visualizations sector-wise
plt.figure(figsize=(12, 6))
```

```python
# Sector-wise line plot
plt.figure(figsize=(15,10))
sns.lineplot(data=df, x='Year', y='General index', hue='Sector')
plt.xlabel('Year')
plt.ylabel('General Index')
plt.title('General Index by Sector')
```

```python
# Sector-wise bar plot
plt.figure(figsize=(15,10))
sns.barplot(data=df, x='Sector', y='General index', hue='Year')
plt.xlabel('Sector')
plt.ylabel('General Index')
plt.title('General Index by Sector and Year')
plt.tight_layout()
plt.show()
```

```python
# Aggregate data for each year and month
total_cereals_by_year_month = df1.groupby(['Year', 'Month'])['Cereals and products'].sum()

# Check if the DataFrame is empty
if total_cereals_by_year_month.empty:
    print("No data available for the given columns.")
else:
    # Reshape the DataFrame for better visualization
    pivot_df = total_cereals_by_year_month.unstack().fillna(0)

    # Plot a pie chart for each year with specified figsize
    fig, axes = plt.subplots(1,len(pivot_df), figsize=(50, 15), subplot_kw=dict(aspect="equal"))
    for i, (year, data) in enumerate(pivot_df.iterrows()):
        ax = axes[i]
        ax.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90)
        ax.set_title(f'Year {year}')
        ax.set_aspect('equal')  # To ensure pie charts are circular

    plt.suptitle('Pie Chart for Cereals and products Distribution by Month (Year-wise)')
    plt.show()
```

```python
# Aggregate data for each year and month
total_cereals_by_year_month = df.groupby(['Year', 'Month'])['Cereals and products'].sum()

# Check if the DataFrame is empty
if total_cereals_by_year_month.empty:
    print("No data available for the given columns.")
else:
    # Reshape the DataFrame for better visualization
    pivot_df = total_cereals_by_year_month.unstack().fillna(0)

    # Get the number of rows and columns for subplots
    num_rows = (len(pivot_df) // 4) + (len(pivot_df) % 4 > 0)  # Calculate the number of rows needed
    num_cols = min(4, len(pivot_df))  # 3 columns or the total number of years, whichever is smaller

    # Plot a pie chart for each year with specified figsize
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15,15))
    for (year, data), ax in zip(pivot_df.iterrows(), axes.flatten()):
        ax.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90)
        ax.set_title(f'Year {year}')
        ax.set_aspect('equal')  # To ensure pie charts are circular
    plt.axis('off')
    plt.subplots_adjust(top=0.75)
    plt.title('Yearwise growth distribution')
    plt.show()
```
